<b> <h3> Bar Graph

A bar graph is a good choice when you want the variation between quantities of a set of items.

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Ghandi", "West Side Story"]
num_oscars = [5,11,3,8,10]               #thats the data
xs = [i  for i,_ in enumerate(movies)]
```

```python
#bars have the default size of 0.8, so we will add 0.1
#to the left coordinates to each bar be centralized.


plt.bar(xs, num_oscars)

plt.ylabel("Number of Prizes")
plt.title("My favorite films")

plt.xticks(xs, movies)

plt.show()
```

    Another example, where the histogram carry numerical values.

```python
from collections import Counter
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade //10 * 10 #it rounds the grades in intervals like 50-60-70...
histogram = Counter(decile(grade) for grade in grades) #it counts the occurancys of grades
histogram
```

```python
plt.bar([x  for x in histogram.keys()], #makes the list
       histogram.values(), #give them the right size
       8) #the width of 8 to each bar

plt.axis([-5,105,0,5])  #-5 to 105 because in the way we can see the full width of the max and minimun's bar

plt.xticks([ i for i in range(0,101,10)])

plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.title("Distribution of Grades")
plt.show()
```

<b> <h3> Line Graphics

```python
variance = [2**x for x in range(9)]
bias_squared = [2**x for x in range(8,-1,-1)]
total_error = [x + y for x,y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
#we can make multiple calls for plt.plot
plt.plot(xs, variance, 'g-', label = 'variance') #solid green line
plt.plot(xs, bias_squared, 'r-.', label = 'bias^2')#red dashed dot line
plt.plot(xs, total_error, 'b:', label = 'total error')#blue dotted line

plt.legend(loc=9) #loc=9 means 'top center'
plt.xlabel("model complexity")
plt.title("Commitment between polarization and variance")
plt.show()
```

<b> <h3> Dispersion Graphics

The best way to visualize the relationship between pairs of data. For example, see the relation between number of friends of each user with the time that they spend in the website.

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = list('abcdefghi')

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, #the name of each point
        xy=(friend_count, minute_count), #the position (x,y) of the point
        xytext=(5,-5), #the position (x,y) of the text using the point as reference
        textcoords='offset points')  #the coordinates system that xytext uses
plt.title("Minutes per Day vs. Number of Friends")
```

We are going to see an example that we have to be careful of,that how matplotlib can mess up with data comprehension when it chooses for us the axis scales.

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("The axis are compatible")
plt.xlabel("grade of test 1")
plt.ylabel("grade of test 2")
plt.show()
```

The upper graph gives the impression that the variance of test one is way bigger than the second test, when in fact that is not true. We can correct that by using <i> plt.axis("equal") </i> and have the true impression from the data.

```python
plt.scatter(test_1_grades, test_2_grades)
plt.title("The axis are compatible")
plt.axis("equal")
plt.xlabel("grade of test 1")
plt.ylabel("grade of test 2")
plt.show()
```
